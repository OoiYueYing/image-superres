```
from PIL import Image
 
#read image
img = Image.open("/content/WAC_TIO2_E350N3150.TIF")
 
#view the image 
#img.show()
```

```
print("Dimensions: {}".format(img.size))
```

```
# Resize image to smaller format i.e: width=600 

basewidth = 800

wpercent = (basewidth/float(img.size[0]))
hsize = int((float(img.size[1])*float(wpercent)))
img_sm = img.resize((basewidth,hsize), Image.ANTIALIAS)

print("Dimensions: {}".format(img_sm.size))
img_sm
```
